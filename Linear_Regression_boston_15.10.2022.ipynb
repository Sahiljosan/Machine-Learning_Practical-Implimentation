## Splitting the data into Train dataset and Test dataset
- We need to do this because once we train the model, the model will not have any information about the test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.33, random_state = 10)
```

- X_train output is y_train
- X_test output is y_test
- X : Independent Feature
- Y : Dependent Feature
- test_size : How much percentage of the data needs to go to test datas
- random_state : Train_test_split that is done, it will be similar

- Train data : use to train the data 
- Test data : use to validate the data, to check how the model is performing well

- X_test and y_test we will not use now. we will use it once the model get trained

### Standardize or Feature Scaling the datasets

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

- `StandardScaler`: It centralize the data, so mean will be zero and there only the distribution will happen
- `Normalization` : In Normalization , u give the range values between 0 to 1 or 1 to 2

```python
scaler
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
# we have use Transform instead of fit_transform to avoid data leakage
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

### Lets train the training data with linear regression

```python
regression.fit(X_train,y_train)
# X_train ----> Independent feature
# y_train ----> Dependent feature
```

- fit() : means train to find out the coefficient
- fit_transform() : means send the data and change the data
- fit_transform() is used because we needed to apply one formula on the entire variables and change the data

- fit_transform() : we are applying the formula and changing the data
- fit() : we are applying the formula but we are not changing the data
- fit() : we are trying to find out the parameters

- means if all the independent features are 0, then the price of the house will be 22.077 (According to the model)

### Print the coefficients and the intercept

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

### Prediction for the test data

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

### Assumptions of Linear Regression
Assumption is used to check whether our linear regression model is good or bad

`Step1` We try to scatter between the real test data and predicted data from the model

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

### Residuals (Errors)
- The difference between y_test and reg_pred

```python
residuals = y_test - reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind = "kde")
```

- The graph is guassian distributed
- Little bit skewed on the right because of the outliers
- Normally whenever we plot distanceplot for residuals, it should approximately be comming like the normal distribution

### Scatter plot with predictions and residual
### Uniform distribution

```python
plt.scatter(reg_pred, residuals)
```

- This distribution is called `uniform distribution`, because all the scatter points are randomly distributed
- Points are randomly distributed,  then also we can say that model is good

### Assumption for Linear Regression
- `1:` when the scatterplot of real test data and regresson prediction shows approximately linear graph
- `2:` whenever we plot distanceplot (displot) for residuals, it should approximately be comming like the normal distribution
- `3:` When the scatter points of regression prediction and residuals are randomly distributed, then we can say that model is good

- This technique is called as `Homoscedasticity`

### Performance Matrix

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

### R square and Adjusted R square

```python
# R Square
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

### Adjusted R square

```python
# Display Adjusted R square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

`Note` Adjusted R square is always less then R square
